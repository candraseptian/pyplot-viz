```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('Downloads/Arsenal WLD.csv')
```

```python
df
```

```python
# Set up the matplotlib figure
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8,6), sharex=True)

# Add a title to the figure
plt.suptitle("Arsenal's Performance Under Arteta Across Seasons", fontsize=16)
# Add a subtitle
plt.figtext(0.5,0.93, "Comparing Victories, Draws, and Defeats Over Time | Viz by @chandraseptiaan | Data via FBRef as of 12th of May 2024",
            fontsize=8, ha='center')

# Generate first Win graph
sns.barplot(x='Season', y='Count', data=df[df['Result'] == 'Win'], color="#00943b", ax=ax1)
ax1.axhline(0, color="k")
ax1.set_ylim(0,31)
ax1.set_ylabel("Win")

# Add value labels for Win graph
for p in ax1.patches:
    ax1.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 5),
                 textcoords = 'offset points',
                 fontsize=10)

# Generate second Draw graph
sns.barplot(x='Season', y='Count', data=df[df['Result'] == 'Draw'], color="#a8a8a8", ax=ax2)
ax2.axhline(0, color="k")
ax2.set_ylim(0, 17)
ax2.set_ylabel("Draw")

# Add value labels for Draw graph
for p in ax2.patches:
    ax2.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 5), 
                 textcoords = 'offset points',
                 fontsize=10)

# Generate third Lose graph
sns.barplot(x='Season', y='Count', data=df[df['Result'] == 'Lose'], color="#e74c3c", ax=ax3)
ax3.axhline(0, color="k")
ax3.set_ylim(0,16)
ax3.set_ylabel("Lose")

# Add value labels for Lose graph
for p in ax3.patches:
    ax3.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center',
                 xytext = (0, 5), 
                 textcoords = 'offset points',
                 fontsize=10)

# Remove x-axis label
ax1.set_xlabel("")
ax2.set_xlabel("")
ax3.set_xlabel("")

# Remove y-axis ticks
ax1.set_yticks([])
ax2.set_yticks([])
ax3.set_yticks([])

# Finalize the plot
sns.despine(bottom=True)
plt.tight_layout(h_pad=2)

plt.savefig('arsenal_performance.png', dpi=300)

plt.show()
```
